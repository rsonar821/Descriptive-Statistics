## **1. Mean**

In statistics, the mean is one of the measures of central tendency, apart from the mode and median. Mean is nothing but the average of the given set of values. It denotes the equal distribution of values for a given data set. To calculate the mean, we need to add the total values given in a dataset and divide the sum by the total number of values.

```python
#Calculating mean from the dataset
print(f"The mean of the monthly income is {df['monthly_income'].mean()} \n")
print(f"The mean of the monthly expense is {df['monthly_expense'].mean()} \n")
print(f"The mean of the monthly rent/EMI amount is {df['rent_amount'].mean()} \n")
print(f"The mean of the annual income is {df['annual_income'].mean()}")
```
